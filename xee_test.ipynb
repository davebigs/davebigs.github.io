```python
import xarray as xr
import ee
```

```python
ee.Authenticate()
ee.Initialize(project = 'ee-dbigelow14')
```

```python
ic = ee.ImageCollection('ECMWF/ERA5_LAND/DAILY_AGGR').filterDate(
    '2023-01-01', '2023-12-31').select('temperature_2m',)

region = ee.Geometry.Rectangle(-130, 25, -60, 50)
ds = xr.open_dataset(ic,engine='ee',crs='EPSG:4326',geometry=region)
ds
```
